```python
corp = """Hello everyone welcome to ICT AI Training.
I am Amey Pilankar."""
print(corp)
## Tokenization

from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt_tab')

document= sent_tokenize(corp)##tokenization of core
type(document)
for sentence in document:
  print(sentence)

from nltk.tokenize import word_tokenize
word_tokenize(corp)
for sentence in document:
  print(word_tokenize(sentence))
```

```python
word=['running','run','walking','codes','coding','programming','completing','completed','science','charul','charuled']
```

```python
from nltk.stem import RegexpStemmer
regexp_stemmer =RegexpStemmer("ing$|ed$|s$",min=3)
for i in word:
  print(regexp_stemmer.stem(i))
```

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
for i in word:
  print(i +" ->- "+ stemming.stem(i))
```

```python
#lemmatize
from nltk.stem import WordNetLemmatizer
Lemmatizer = WordNetLemmatizer()
for i in word:
  print(i +" ->- "+ Lemmatizer.lemmatize(i,pos='v'))
```
