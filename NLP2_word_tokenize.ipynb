```python
from nltk.tokenize import word_tokenize
sentence="798 89 9 00 charul is a good boy, he is working very hardworking coding coder completely"
words= word_tokenize(sentence)
print(words,type(words))
```

```python
pos_tags = nltk.pos_tag(words)
for words, tag in pos_tags:
  print(words,"-",tag)
```

```python
from nltk.stem import RegexpStemmer
regexp_stemmer =RegexpStemmer("ing$|ed$|s$",min=3)
for i in words:
  print(regexp_stemmer.stem(i))
```
